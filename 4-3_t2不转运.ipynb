```python
raw_order=pd.read_excel(r"D:\数模\第四次模拟题——在线零售商的订单履行\附件一.xlsx")
raw_warehouse=pd.read_excel(r"D:\数模\第四次模拟题——在线零售商的订单履行\附件二.xlsx")
```

```python
SKU1=raw_order['订单中的SKU']
SKU_in_order=[]
# 原有数据类型为'[]'型，先进行转化,最终要得到['','']的形式
for order in SKU1:
    #print(type(order))
    if type(order)==int:
        print(str(order))
        SKU_in_order.append(str(order))
    else:
        order=order[1:-1].split(",")
        SKU_in_order.append(order)
```

```python
SKU_w1=raw_warehouse['仓库1'][:-1]
pos_w1=[0.516205320509211,83.7701245337600]
SKU_w2=raw_warehouse['仓库2'][:-1]
pos_w2=[30.0817907887535,1.17493342829637]
SKU_w3=raw_warehouse['仓库3'][:-1]
pos_w3=[58.8349268443702,89.8949407626723]
```

```python
# 储存货品编号
W=[]
W1=[]
W2=[]
W3=[]
for i in range(len(SKU_w1)):
    if SKU_w1[i]==1:
        W1.append(str(i+1))
    if SKU_w2[i]==1:
        W2.append(str(i+1))
    if SKU_w3[i]==1:
        W3.append(str(i+1))
W1=set(W1)
W2=set(W2)
W3=set(W3)
W.append(W1)
W.append(W2)
W.append(W3)

print(len(W1))
```

```python
pos_order=raw_order['订单对应的顾客位置坐标']
pos_o=[]
for i in pos_order:
    i=i[1:-1].split(",")
    pos_o.append([float(t) for t in i])
```

```python
import math
def dis(order_pos):
    dis1=math.sqrt((order_pos[0]-pos_w1[0])**2+(order_pos[1]-pos_w1[1])**2)
    dis2=math.sqrt((order_pos[0]-pos_w2[0])**2+(order_pos[1]-pos_w2[1])**2)
    dis3=math.sqrt((order_pos[0]-pos_w3[0])**2+(order_pos[1]-pos_w3[1])**2)
    return [dis1,dis2,dis3]
```

```python
cost=[]
package=0
cm1=0
cm2=0
cm3=0
for i in range(len(SKU_in_order)):
    #single_cost=0
    if type(SKU_in_order[i])==str:
        k1=int(SKU_in_order[i])
        l1=[str(k1)]
        order_need=set(l1)
    else:
        order_need=set(SKU_in_order[i])

    pos=dis(pos_o[i])
    A=pos.index(min(pos))
    C=pos.index(max(pos))

    for j in pos:
        if j<max(pos) and j >min(pos):
            B=pos.index(j)

    if min(pos)<5:
            costA=3
    else :
            costA=3+0.2*(min(pos)-5.0)

    if pos[B]<5:
            costB=3
    else :
            costB=3+0.2*(pos[B]-5.0)
    
    if max(pos)<5:
            costC=3
    else :
            costC=3+0.2*(max(pos)-5.0)

    single_cost=9999
    if order_need.issubset(W[A]):
        single_cost=min(single_cost,costA)
    elif order_need.issubset(W[B]):
        single_cost=min(single_cost,costB)
    elif order_need.issubset(W[C]):
        single_cost=min(single_cost,costC)

    elif order_need.issubset(W[A]|W[B]):
        single_cost=min(single_cost,costA+costB)
    elif order_need.issubset(W[A]|W[C]):
        single_cost=min(single_cost,costA+costC)
    elif order_need.issubset(W[C]|W[B]):
        single_cost=min(single_cost,costC+costB)
    else:
        single_cost=min(single_cost,costA+costB+costC)
    if single_cost==costA or single_cost==costB or single_cost==costC:
        cm1+=1
    if single_cost==costA+costB or single_cost==costA+costC or single_cost==costB+costC:
        cm2+=1
    if single_cost==costA+costB+costC:
        cm3+=1
    cost.append(single_cost)

```
